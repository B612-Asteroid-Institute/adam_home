```python
import numpy as np
import pandas as pd

from stk.io import convertPointingsToVectorInterval
```

Reduced OpSim files:
- [1 Year](https://github.com/moeyensj/impactProbabilityStudy/blob/master/opsimReduced/minion_1016_reduced_1y.txt)
- [5 Year](https://github.com/moeyensj/impactProbabilityStudy/blob/master/opsimReduced/minion_1016_reduced_5y.txt)
- [10 Year](https://github.com/moeyensj/impactProbabilityStudy/blob/master/opsimReduced/minion_1016_reduced_10y.txt)

To create arbitrary length OpSim files see this repo: [impactProbabilityStudy](https://github.com/moeyensj/impactProbabilityStudy/)



```python
# Assumes reduced file is in notebook directory
reducedOpsimFile = "minion_1016_reduced_10y.txt"
pointings = pd.read_csv(reducedOpsimFile, sep=" ") 
# might have to add in index=1 
# as a keyword argument here if the opsim file has the index printed 
```

```python
pointings
```

```python
convertPointingsToVectorInterval("VectorFile", "IntervalList",
                                pointings["expDate"].values,
                                np.degrees(pointings["fieldRA"].values),
                                np.degrees(pointings["fieldDec"].values),
                                epochStart = "1 Jan 2022 00:00:00",
                                exposureLength = 30)
```
